```python
from scrapy.selector import Selector
import pandas as pd
```

```python
driver =  webdriver.Chrome(executable_path="D:/Tamil/chrome driver/chromedriver.exe")
driver.get('https://www.remax.ca/real-estate-agents-in-canada')
sleep(2)
expander = driver.find_elements_by_xpath('//span[@class="material-icons"]')
print(len(expander))
for exp in expander:
    sleep(2)
    exp.click()
html = driver.page_source
resp = Selector(text=html)
print(resp)
links = resp.xpath('//div[@class="link-card-container"]/a/@href').getall()

#Extracting Office Links
z= 0
df = pd.DataFrame([])
for link in links:
    df.loc[z,'links'] = 'https://www.remax.ca'+str(link)
    
    z = z+1

     
```

```python
driver.get(office_link['links'][0])
z = 0
agent_links_df = pd.DataFrame([])
for page in range(100):
    page_link = 'https://www.remax.ca/on/toronto-real-estate-agents?pageNumber='+str(page+1)
    driver.get(page_link)
    sleep(1)
    agent_links = driver.find_elements_by_xpath('//*[@class="card-with-buttons_content___GK4m"]/a')
    for link in agent_links:
        #rint(link.get_attribute('href'))
        agent_links_df.loc[z,'links']= link.get_attribute('href')
        z = z+1
```

```python
agent_links_df.to_csv('remax_ca_links.csv',index=False)
```
