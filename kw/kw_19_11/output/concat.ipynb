```python
import pandas as pd
pd.set_option('display.max_columns', 65)
```

```python
import os
import glob
#directory
path = os.getcwd() 

#getting All csv file names from the directory
csv_files = glob.glob(os.path.join(path, "*.csv"))
```

```python
csv_files
```

```python
overall_op = pd.read_csv(csv_files[0])
for i in range(len(csv_files)):
    if i != 0:
        a = pd.read_csv(csv_files[i])
        overall_op = pd.concat([overall_op,a])
```

```python
overall_op.reset_index(drop= True, inplace= True)
```

```python
overall_op = overall_op[overall_op['agent_name'].notna()]
```
