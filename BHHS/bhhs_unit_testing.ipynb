```python
office_info = pd.DataFrame([])
z = 0

for page in range(0,30):
    

    soup = BeautifulSoup(driver.page_source,'html.parser')
    all_office_details = soup.find_all('div',{'class':'row office-list pt-3 pb-2'})
    for office_details in all_office_details:
        try:
            office_url = ''.join([x['href'] for x in office_details.find_all('a',{'class':'btn btn-secondary'}) if 'contact-us' not in x['href']])
            if 'http' not in office_url:
                office_url = 'https://www.bhhs.com'+office_url
        except:
            office_url = ''
            
        try:
            office_name = office_details.find('div',{'class':'col-12 col-lg-6'}).text
        except:
            office_name = ''
            
        try:
            office_address = office_details.find('div',{'class':'office-list__location-address b2'}).text.replace('\n                           ','').replace('\n','').replace('  ',' ').strip()
        except:
            office_address = ''
            
        try:
            office_phone = ''
            office_fax = ''
            for ind , no in enumerate(office_details.find_all('div',{'class':'number'})):
                if ind == 0:
                    office_phone = no.text
                if ind == 1:
                    office_fax = no.text
        except:
            office_phone = ''
            office_fax = ''
            
        office_info.loc[z,'office_site'] = office_url
        office_info.loc[z,'office_name'] = office_name
        office_info.loc[z,'office_phone'] = office_phone
        office_info.loc[z,'fax'] = office_fax
        office_info.loc[z,'address'] = office_address
        
        z = z+1
        

    try:         
        driver.find_element_by_xpath('//*[@class="cmp-search-results-pagination__arrow cmp-search-results-pagination__arrow--next d-none d-lg-inline-block"]').click()
        time.sleep(5)
    except:
        break
```
