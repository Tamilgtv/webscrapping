```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from datetime import datetime
df = pd.DataFrame([])
z = 0
for pages in range(0,357):
    if pages == 0:
        page = requests.get('https://www.iprorealty.com/agent_find')
    else:
        page = requests.get('https://www.iprorealty.com/agent_find?page='+str(pages))
        
    soup = BeautifulSoup(page.content,'html.parser')
    
    for agent_detail in soup.find_all('div',{'class':'agent-summary-wrapper'}):
        try:
            agent_url = 'https://www.iprorealty.com'+agent_detail.find('a')['href']
        except:
            agent_url = ''
        try:
            agent_name = agent_detail.find('h3',{'class':'agent-name'}).text
        except:
            agent_name = ''
        try:
            agent_dp = 'https://www.iprorealty.com'+agent_detail.find('img')['src']
        except:
            agent_dp = ''
        try:
            agent_role = agent_detail.find('h4',{'class':'agent_title'}).text
        except:
            agent_role = ''
        try:
            agent_mobile = agent_detail.find('a',{'class':'mobile-phone-link'})['href'].replace('tel:','').replace('Direct','').strip() #agent_mobile
        except:
            agent_mobile = ''
        try:
            agent_site = agent_detail.find('a',{'class':'agent-site-link'})['href']
        except:
            agent_site = ''
        try:
            office_phone = agent_detail.find('a',{'class':'office-phone-link'})['href'].replace('tel:','')
        except:
            office_phone = ''
        try:
            office_name = 'iPro Realty Ltd., Brokerage'
        except:
            office_name = ''
        try:
            office_address = agent_detail.find('span',{'class':'brokerage-information contact-data'}).text
        except:
            office_address = ''
            
            
        try:
            state = office_address.split(',')[-2]
        except:
            state = ''
        try:
            Zip = office_address.split(',')[-1]
        except:
            Zip = ''
            
#         df.loc[z,'unique_id'] = 'ipro_realty_'+str(z+1)
#         df.loc[z,'agent_url'] = agent_url
#         df.loc[z,'agent_dp'] = agent_dp
#         df.loc[z,'agent_name'] = agent_name
#         df.loc[z,'agent_role'] = agent_role
#         df.loc[z,'agent_mobile'] = agent_mobile
#         df.loc[z,'agent_site'] = agent_site
#         df.loc[z,'office_phone'] = office_phone
#         df.loc[z,'office_name'] = office_name
#         df.loc[z,'office_address'] = office_address
#         z = z+1
        
        
        
        
        df.loc[z,'unique_id'] = 'ipro_realty_'+str(z+1)
        df.loc[z, 'agent_url'] = agent_url
        df.loc[z, 'agent_dp'] = agent_dp
        df.loc[z, 'agent_name'] = agent_name
        df.loc[z, 'agent_phone'] = agent_mobile
        df.loc[z, 'agent_email'] = ''
        df.loc[z, 'agent_role'] = agent_role
        df.loc[z, 'agent_license'] = ''
        df.loc[z, 'agent_areaserved'] = ''
        df.loc[z, 'agent_designations'] = ''
        df.loc[z, 'agent_facebook'] = ''
        df.loc[z, 'agent_instagram'] = ''
        df.loc[z, 'agent_twitter'] = ''
        df.loc[z, 'agent_linkedin'] = ''
        df.loc[z, 'agent_pinterest'] = ''
        df.loc[z, 'agent_youtube'] = ''
        df.loc[z, 'agent_site'] = agent_site
        df.loc[z, 'agent_special_tag'] = ''
        df.loc[z, 'agent_rating'] = ''
        df.loc[z, 'office_address'] = office_address
        df.loc[z, 'office_fax'] = ''
        df.loc[z, 'office_branch'] = ''
        df.loc[z, 'office_name'] = office_name
        df.loc[z, 'office_phone'] = office_phone.replace('+1','')
        df.loc[z, 'office_site'] = ''
        df.loc[z, 'office_zip'] = Zip
        df.loc[z, 'office_state'] = state
        df.loc[z,'company_facebook'] = ''
        df.loc[z,'company_linkedin'] = ''
        df.loc[z,'company_name'] = 'Ipro Realty'
        df.loc[z,'company_site'] = 'https://www.iprorealty.com'
        df.loc[z,'company_twitter'] = ''
        df.loc[z,'date_of_data_extraction'] = datetime.now()
        df.loc[z,'data_extracted_by'] = 'Tamil'
        df.loc[z,'dataextraction_uuid'] = 'Tamil-'+datetime.now().strftime('%m-%d-%Y')

        z = z+1
```
