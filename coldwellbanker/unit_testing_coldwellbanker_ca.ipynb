```python
import requests
```

```python
page = requests.get('https://www.coldwellbanker.ca/agents/ca/?sort=relevance&query_type=all-agents&ExpandQuery=true')
```

```python
import pandas as pd
```

```python
page = ''
z = 0
agent_links = pd.DataFrame([])
for pages in range(1,216):
    print(pages)

    if pages == 1:
        page = requests.get('https://www.coldwellbanker.ca/agents/ca/?sort=relevance&query_type=all-agents&ExpandQuery=true')
        #driver.get('https://www.coldwellbanker.ca/agents/ca/?sort=relevance&query_type=all-agents&ExpandQuery=true')
    else:
        page = requests.get('https://www.coldwellbanker.ca/agents/ca/page/'+str(pages)+'/?sort=name&query_type=all-agents&ExpandQuery=true')
    soup = BeautifulSoup(page.content,'html.parser')
    for agent_details in soup.find_all('div',{'class':'agent-preview-meta'}):
        
        agent_url = agent_details.find('a')['href']
        agent_links.loc[z,'links'] = agent_url
        z = z + 1
        
```

```python
agent_links.to_csv('All_agent_ca.csv',index=False)
```

```python
df = pd.DataFrame([])
z = 0
for link in agent_links['links']:
    print(z)
    
    page = requests.get(link)
    soup = BeautifulSoup(page.content,'html.parser')
    #agent_details = soup.find_all('div',{'class':'agent-preview-meta'})
    agent_details = soup.find('div',{'class':'agent-preview-meta'})
    try:
        agent_dp = soup.find('a',{'class':'profile-image'}).find('img')['data-lazy-src']
    except:
        agent_dp = ''
    try:
        name = agent_details.find('h1').text.strip()
    except:
        name = ''
    try:
        agent_url = link
    except:
        agent_url = ''
    try:
        agent_role = agent_details.find('span',{'class':'job-title'}).text.strip()
    except:
        agent_role = ''
    try:
        office_name = agent_details.find('address',{'class':'agent-office'}).find('span').text.strip()
    except:
        office_name = ''
    try:
        office_site = agent_details.find('address',{'class':'agent-office'}).find('span').find('a')['href']
    except:
        office_site = ''
    try:

        address = ''.join([x.text for x in agent_details.find('address',{'class':'agent-office'}).find_all('span')][1:2]).replace('#','').strip()
        state = address.split(' ')[-2]
        Zip = address.split(' ')[-1]
    except:
        address = ''
        state = ''
        Zip = ''
    try:
        mobile = ''.join([x.text for x in agent_details.find_all('p',{'class':'agent-line'}) if 'Mobile' in x.text ])
    except:
        mobile = ''

    try:
        office = ''.join([x.text for x in agent_details.find_all('p',{'class':'agent-line'}) if 'Office' in x.text ])
    except:
        office = ''

    try:
        agent_site = ''.join([x.find('a')['href'] for x in agent_details.find_all('p',{'class':'agent-line'}) if 'Web' in x.text ])
    except:
        agent_site = ''

    try:

        agent_email = ''.join([x['href'] for x in agent_details.find('div',{'class':'agent-contacts-actions'}).find_all('a') if 'mailto' in x['href'] ]).replace('mailto:','').strip()
    except:
        agent_email = ''
    try:

        agent_facebook = ''.join([x['href'] for x in agent_details.find('div',{'class':'agent-contacts-actions'}).find_all('a') if 'facebook' in x['href'] ])
    except:
        agent_facebook = ''
    try:
        agent_linkedin = ''.join([x['href'] for x in agent_details.find('div',{'class':'agent-contacts-actions'}).find_all('a') if 'linkedin' in x['href'] ])
    except:
        agent_linkedin = ''

    try:
        agent_insta = ''.join([x['href'] for x in agent_details.find('div',{'class':'agent-contacts-actions'}).find_all('a') if 'insta' in x['href'] ])
    except:
        agent_insta = ''

    try:
        agent_twitter = ''.join([x['href'] for x in agent_details.find('div',{'class':'agent-contacts-actions'}).find_all('a') if 'twitter' in x['href'] ])
    except:
        agent_twitter = ''



    df.loc[z,'unique_id'] = 'coldwellbanker_ca_'+str(z+1)
    df.loc[z, 'agent_url'] = link
    df.loc[z, 'agent_dp'] = agent_dp
    df.loc[z, 'agent_name'] = name
    df.loc[z, 'agent_phone'] = mobile.replace('Mobile:','').replace('+1','').strip()
    df.loc[z, 'agent_email'] = agent_email
    df.loc[z, 'agent_role'] = agent_role
    df.loc[z, 'agent_license'] = ''
    df.loc[z, 'agent_areaserved'] = ''
    df.loc[z, 'agent_designations'] = ''
    df.loc[z, 'agent_facebook'] = agent_facebook
    df.loc[z, 'agent_instagram'] = agent_insta
    df.loc[z, 'agent_twitter'] = agent_twitter
    df.loc[z, 'agent_linkedin'] = agent_linkedin
    df.loc[z, 'agent_pinterest'] = ''
    df.loc[z, 'agent_youtube'] = ''
    df.loc[z, 'agent_site'] = agent_site
    df.loc[z, 'agent_special_tag'] = ''
    df.loc[z, 'agent_rating'] = ''
    df.loc[z, 'office_address'] = address
    df.loc[z, 'office_fax'] = ''
    df.loc[z, 'office_branch'] = ''
    df.loc[z, 'office_name'] = office_name
    df.loc[z, 'office_phone'] = office.replace('+1','')
    df.loc[z, 'office_site'] = office_site
    df.loc[z, 'office_zip'] = Zip
    df.loc[z, 'office_state'] = state
    df.loc[z,'company_facebook'] = ''
    df.loc[z,'company_linkedin'] = ''
    df.loc[z,'company_name'] = 'Coldwellbanker'
    df.loc[z,'company_site'] = 'https://coldwellbanker.ca'
    df.loc[z,'company_twitter'] = ''
    df.loc[z,'date_of_data_extraction'] = datetime.now()
    df.loc[z,'data_extracted_by'] = 'Tamil'
    df.loc[z,'dataextraction_uuid'] = 'Tamil-'+datetime.now().strftime('%m-%d-%Y')

    z = z+1
```
