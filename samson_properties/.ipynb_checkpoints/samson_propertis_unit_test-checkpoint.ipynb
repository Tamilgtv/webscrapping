```python
from selenium import webdriver
from datetime import datetime
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
driver = webdriver.Chrome(executable_path='D:/Tamil/chrome driver/chromedriver.exe')
```

```python
driver.find_element_by_xpath('//*[@class = "smallVPad"]/strong/a').text
```

```python
driver.get('https://www.samsonproperties.net/agents.php')
```

```python
driver.find_element_by_xpath('//*[@class = "listing-box-title"]/h2/a').get_attribute('href')
```

```python
df = pd.DataFrame([])
z = 0
for page in range(1, 11):
    driver.get('https://www.samsonproperties.net/agents.php?page='+str(page))
    for no, agent_link in  enumerate(driver.find_elements_by_xpath('//*[@class = "listing-box-title"]/h2/a')):
        print(z)
        df.loc[z,'links'] = agent_link.get_attribute('href')
        z = z+ 1
        #print(no,agent_link.get_attribute('href'))
```

```python
agent_links = df.copy()
```

```python
df = pd.DataFrame([])
z = 0
for link in agent_links['links'][:250]:
    driver.get(link)
    print(z)
    
    agent_url = link
    try:
        agent_dp = driver.find_element_by_xpath('//*[@class = "listing-user"]/div/a').get_attribute('style').replace('background-image: url("','').replace('");','')
    except:
        agent_dp = ''
    try:
        agent_name = ''.join([x.text for x in driver.find_elements_by_xpath('//*[@class="overview"]/ul/li') if 'Name' in x.text]).replace('Name\n','').strip()
    except:
        agent_name = ''
    try:
        agent_role = driver.find_element_by_tag_name('h1').text.split('-')[-1].strip()
    except:
        agent_role = ''
    
    try:
        agent_site = driver.find_element_by_xpath('//*[@class="overview"]/ul/li/strong/a').get_attribute('href')
    except:
        agent_site = ''
        
    try:
        office_name = driver.find_element_by_xpath('//*[@class = "listing-user-title"]/h3').text
    except:
        office_name =''
        
    try:
        agent_phone = ''.join([x.text for x in driver.find_elements_by_xpath('//*[@class="overview"]/ul/li') if 'Cell' in x.text]).replace('Cell\n','')
    except:
        agent_phone = ''
        
    try:
        other_phone = ''.join([x.text for x in driver.find_elements_by_xpath('//*[@class="overview"]/ul/li') if 'Work' in x.text]).replace('Work Phone\n','')
    except:
        other_phone = ''
        
    try:
        agent_facebook = ''.join([x.get_attribute('href') for x in driver.find_elements_by_xpath('//*[@class="listing-user-social"]/ul/li/a') if 'face' in x.get_attribute('href')])
    except:
        agent_facebook = ''
        
    try:
        agent_instagram = ''.join([x.get_attribute('href') for x in driver.find_elements_by_xpath('//*[@class="listing-user-social"]/ul/li/a') if 'insta' in x.get_attribute('href')])
    except:
        agent_instagram = ''
        
        
    try:
        agent_twitter = ''.join([x.get_attribute('href') for x in driver.find_elements_by_xpath('//*[@class="listing-user-social"]/ul/li/a') if 'twitter' in x.get_attribute('href')])
    except:
        agent_twitter = ''
        
    try:
        agent_linkedin = ''.join([x.get_attribute('href') for x in driver.find_elements_by_xpath('//*[@class="listing-user-social"]/ul/li/a') if 'linkedin' in x.get_attribute('href')])
    
    except:
        agent_linkedin = ''
        
        
    try:
        office_address = ''.join([x.text for x in driver.find_elements_by_xpath('//*[@class="contact"]/tbody/tr') if 'Address' in x.text]).replace('Address:','').replace('\n',' ').strip()
        state = office_address.split(' ')[-2]
        Zip = office_address.split(' ')[-1]
    except:
        office_address = ''
        state = ''
        Zip = ''

    df.loc[z, 'agent_url'] = link
    df.loc[z, 'agent_dp'] = agent_dp
    df.loc[z, 'agent_name'] = agent_name
    df.loc[z, 'agent_phone'] = agent_phone
    df.loc[z, 'agent_email'] = ''
    df.loc[z, 'agent_role'] = agent_role
    df.loc[z, 'agent_license'] = ''
    df.loc[z, 'agent_areaserved'] = ''
    df.loc[z, 'agent_designation'] = ''
    df.loc[z, 'agent_facebook'] = agent_facebook
    df.loc[z, 'agent_instagram'] = agent_instagram
    df.loc[z, 'agent_twitter'] = agent_twitter
    df.loc[z, 'agent_linkedin'] = agent_linkedin
    df.loc[z, 'agent_pinterest'] = ''
    df.loc[z, 'agent_youtube'] = ''
    df.loc[z, 'agent_site'] = agent_site
    df.loc[z, 'agent_special_tag'] = ''
    df.loc[z, 'agent_rating'] = ''
    df.loc[z, 'office_address'] = office_address
    df.loc[z, 'office_fax'] = ''
    df.loc[z, 'office_branch'] = office_name
    df.loc[z, 'office_name'] = 'Samson Properties'
    df.loc[z, 'office_phone'] = ''
    df.loc[z, 'office_site'] = ''
    df.loc[z, 'office_zip'] = Zip
    df.loc[z, 'office_state'] = state
    df.loc[z,'other_phone'] = other_phone
    df.loc[z,'company_facebook'] = ''
    df.loc[z,'company_linkedin'] = ''
    df.loc[z,'company_name'] = 'Samson Properties'
    df.loc[z,'company_site'] = 'https://www.samsonproperties.net'
    df.loc[z,'company_twitter'] = ''
    df.loc[z,'date_of_data_extraction'] = datetime.now()
    df.loc[z,'data_extracted_by'] = 'Tamil'
    df.loc[z,'dataextraction_uuid'] = ''

    z = z+1

df.to_csv('samson_properties_sample.csv',index=False)
        
```
